```julia
using Images, Plots
using LinearAlgebra
```

```julia
f1(x, y) = -2*x*y
f2(x, y) = (x^2-y^2)
```

```julia
xs = -4:0.5:4
ys = -4:0.5:4

df(x, y) = normalize([f1(x, y), f2(x, y)])./2.0

xxs = [x for x in xs for y in ys]
yys = [y for x in xs for y in ys]

quiver(xxs, yys, quiver=df, frame = :origin, aspect_ratio = :equal, size = (400, 400))
#savefig("problem_13_1_1.png")
```

```julia
t = 0.0:0.01:2

xx = cos.(t)
yy = sin.(3.0 .* t)

plot(xx, yy, frame = :origin)
#savefig("problem_13_3_1_b.png")
```

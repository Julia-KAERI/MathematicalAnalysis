```julia
using CairoMakie, LaTeXStrings
```

```julia
x = 1.0:0.01:3.0

```

```julia
α=3.0
f(x) = 0.5*(x+α/x)
g(x) = (α+x)/(1+x)

fp(x) = 0.5- α/2/x/x
gp(x) = -(α-1)/((1+x)^2);
fig0 = Figure()
ax0 = Axis(fig0[1, 1])

lines!(ax0, x, f.(x), label = L"f(x)")
lines!(ax0, x, g.(x), label = L"g(x)")
lines!(ax0, x, x, label = L"y=x")
axislegend(position = :rb)
fig0
save("pp1.png", fig0)
```
